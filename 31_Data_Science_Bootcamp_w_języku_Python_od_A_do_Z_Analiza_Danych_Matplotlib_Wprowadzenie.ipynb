#Import bibliotek

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
````

````python
#nałożenie stylu z biblioteki Seaborn na nasze wykresy
sns.set()
````

````python
fig = plt.figure()
fig.show()
````

# Wykres liniowy

````python
plt.plot([0, 1], [0, 3])
````

````python
plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color = 'g')
plt.legend()
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('wykres liniowy')
plt.show
````

Narysujmy sobie funkcję Sinus. Aby to zrobić musimy wykorzystać w pierwszej kolejności bibliotekę numpy 

````python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
````

````python
plt.figure(figsize = (10,6))
plt.plot(x, x_sin, marker = 'o')
plt.plot(x, x_cos)
plt.show()
````

#Tworzenie wielu wykresów

Stwórzmy sobie dwa wykresy na jednym obrazku

````python
plt.figure(figsize = (12,6))
plt.subplot(121) # 1 - liczba wierszy w naszej siatce, 2 - liczba kolumn w naszej siatce, 1 - odniesienie do konkretnego wykresu
plt.plot([0, 1, 2], [3, 2, 5], linestyle = ':' )

plt.subplot(122)
plt.plot([4, 5, 6], [3, 7, 2], linestyle = '--', color = 'r')
plt.show()
````

Stwórzmy sobie 3 wykresy

````python
plt.figure(figsize = (8,10)) #szerokośc i wysokość - zwracane w calach
plt.subplot(311) # 3 - liczba wierszy w naszej siatce, 1 - liczba kolumn w naszej siatce, 1 - odniesienie do konkretnego wykresu
plt.plot([0, 1, 2], [3, 2, 5], linestyle = ':' )

plt.subplot(312)
plt.plot([4, 5, 6], [3, 7, 2], linestyle = '--', color = 'r')

plt.subplot(313)
plt.plot([7, 2, 4], [5, 9, 2], marker = 'o', linestyle = 'None', color = 'g')

plt.show()
````

#Wykresy słupkowe

````python
plt.bar(x = [0, 1, 2, 3], height = [1.2, 3.2, 1.6, 2.3])
plt.show()
````

````python
plt.bar(x = ['a', 'b', 'c', 'd'], height = [1.2, 3.2, 1.6, 2.3])
plt.show()
````

````python
#wykres horyzontalny
plt.barh(y = ['a', 'b', 'c', 'd'], width = [1.2, 3.2, 1.6, 2.3], color = 'r')
plt.show()
````

#Wykres punktowy

````python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
````

````python
plt.scatter(x = x1, y = x2, color = 'r')
````

Różna wielkośc kropek na naszym wykresie punktowym - musimy dodać jeszcze jedną zmienną (np. x3)

````python
x3 = np.random.randn(300) * 50
plt.scatter(x = x1, y = x2, color = 'tab:purple', s = x3, alpha = 0.6) #parametr alpha - przezroczystość, parametr 's' - wielkosć kropek
_ = plt.show()
````
