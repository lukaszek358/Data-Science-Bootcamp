## Import bibliotek

````python
import pandas as pd
import numpy as np
````

#Załadowanie danych

````python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')
````

````python
df = fetch_financial_data('AMZN')
df.head()
````

#Zapis/Odczyt danych z pliku CSV

````python
df.to_csv('Amazon.csv')
````

Wyciągamy dane za miesiąc listopad na 2019 rok

````python
df_nov_2019 = df[(df.index.month == 11) & (df.index.year == 2019)]
df_nov_2019
````

Zapisujemy ramkę danych df_nov_2019 do pliku

````python
df_nov_2019.to_csv('Amazon_2019_11.csv')
````

Odczyt naszych danych

````python
df_new = pd.read_csv('Amazon_2019_11.csv', index_col = 0)
df_new
````

Zapis do pliku excel jako .xlsx

````python
df_nov_2019.to_excel('Amazon_2019_11.xlsx')
````

Odczyt pliku z excela

````python
df_new = pd.read_excel('Amazon_2019_11.xlsx', index_col = 0) #kolumna Date jako indeks kolumny
df_new.head()
````
