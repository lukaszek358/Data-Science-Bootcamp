#Import bibliotek


````python
import pandas as pd
import numpy as np
````

#Pobieranie danych

````python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')
````

Określone bazy danych dla spółek notowanych na giełdzie zapisujemy jako zmienne (nazwa spółki)

````python
apple = fetch_financial_data('AAPL')
amazon = fetch_financial_data('AMZN')
google = fetch_financial_data('GOOGL')
uber = fetch_financial_data('UBER')
````

#Preprocesing

Przetwarzamy nazwy naszych kolumn

````python
#dodawanie przedrostka 'apple_' do każdej z kolumn w bazie apple + użyjemy metody .lower aby zmienić nazwy kolumn na małe litery
apple.columns = ['apple_' + col.lower() for col in apple.columns]
````

````python
#robimy tak również dla pozostałych df
amazon.columns = ['amazon_' + col.lower() for col in amazon.columns]

uber.columns = ['uber_' + col.lower() for col in uber.columns]

google.columns = ['google_' + col.lower() for col in google.columns]
````

````python
apple.head()
````

# Konkatenacja danych

W programowaniu oznacza łączenie dwóch wyrażeń (np. tekstowych) w jedno (ustawienie jednego za drugim). 

````python
pd.concat(objs = [apple, amazon, google, uber], axis = 1) #axis = 1 - łączenie po kolumnach/ dodawanie jako kolumny
````

````python
#przypisanie naszych połączonych danych do zmiennej df
df = pd.concat(objs = [apple, amazon, google, uber], axis = 1)

#wyświetlenie opisu danych znajdujących się w tabeli df (wraz z transponowaniem - czytelniejsza wersja)
df.describe().T
````

````python
#zmiana notacji wykładniczej na czytelniejszą, do dwóch miejsc po przecinku

pd.set_option('display.float_format', lambda x: f'{x:.2f}')

df.describe().T
````

````python
#wyświetlenie współczynnika korelacji
df.corr()
````

````python
#zwrócenie wsyztskich dostępnych kolumn w naszym df
df.columns
````

````python
#zwrócenie kolumn zakończonych zwrotem 'close' - list_comprehension

closes = [col for col in df.columns if col.endswith('close')]
closes
````

````python
#df składający się tylko z kolumn zamknięcia
df[closes]
````

````python
#korelacja dla zmiennych z powyższej tabeli
df[closes].corr()
````

````python
#df składający się tylko z kolumn zamknięcia i wolumenu obrotu
closes_volume = [col for col in df.columns if col.endswith('close') or col.endswith('volume')]
df[closes_volume]
````

````python
#korelacja dla zmiennych z powyższej tabeli
df[closes_volume].corr()
````
