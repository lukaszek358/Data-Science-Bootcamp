#Import bibliotek

````python
import numpy as np
import pandas as pd
````

#Pobranie i załadowanie danych

````python
url = 'https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/appstore_games.csv'
#df_raw - 'raw' mówi nam, że nasze dane na obecną chwilę w DataFrame są surowe (nieobrobione/ nieprzygotowane)
df_raw = pd.read_csv(url)
````

````python
df_raw.head()
````

#Wstępne przygotowanie danych

````python
print(df_raw.columns, '\n')
df = df_raw.copy()
df.info()
````

Usuwanie kolumn z naszego DF - usuwamy kolumny, które Nas nie interesują.

````python
df = df.drop(columns = ['URL', 'Icon URL', 'Subtitle', 'Description', 'In-app Purchases'])
````

````python
df.head()
````

````python
#wykorzystanie kolumny ID jako kolumny indeksów
df = df.set_index('ID')
````

````python
df.head()
````

Wyświetlan ie statystyk naszego DF - describe() zwróci nam tylko informacje dla danych numerycznych.

````python
df.describe()
````

W komendzie musimy wykorzystać warunek, który spowoduje wzięcie pod uwagę danych kategorycznych.

````python
df.describe(include = ['object']) #object - typ danych kategorycznych (tekstowych)
````

````python
df.describe(include = ['object']).T
````

## Standaryzacja nazw kolumn

````python
df.columns
````

````python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
df.head()
````

# Badanie rozkładu zmiennych

````python
df.columns
````

````python
df.head(2)
````

````python
#sprawdzamy częstość/liczność wystąpienia każdego obiektu w kolumnie primary_genre
df.primary_genre.value_counts()
````

````python
#zwracamy 5 najbardziej popularnych gatunków

df.primary_genre.value_counts().nlargest(5)
````

````python
#zwracamy 5 najbardziej popularnych gatunków (same nazwy !)
df.primary_genre.value_counts().nlargest(5).index
````

````python
#konwersja powyższych danych do listy
list(df.primary_genre.value_counts().nlargest(5).index)
````

````python
#sprawdzamy częstość/liczność wystąpienia każdego obiektu w kolumnie age_rating - wygenerowane jako lista
list(df.age_rating.value_counts().index)
````

````python
df.age_rating.value_counts()
````

````python
#sprawdzamy częstość/liczność wystąpienia każdego obiektu w kolumnie price
df.price.value_counts()
````

````python
#sprawdzamy częstość/liczność wystąpienia każdego obiektu w kolumnie average_user_rating
df.average_user_rating.value_counts()
````

````python
#badamy kolumnę languages
df.languages
````

Chcemy sprawdzić ile języków znajduje się w aplikacji.
W związku z tym i powyższymi danymi będziemy pracować na tekście - skorzystamy zatem z metody .str oraz metody podziału .split()

````python
#elementy (języki) zostały zwrócone jako lista
df.languages.str.split(', ')
````

````python
#zwracamy długośc każdej listy z językami (ilośc elementów)
df.languages.str.split(', ').str.len()
````

````python
#przypisujemy powyższą wyciągniętą ilość elementów do nowej zmiennej jako kolumna do naszego DataFrame'u
df['num_languages'] = df.languages.str.split(', ').str.len() #liczba języków dostępna w każdej aplikacji 
df.head()
````

````python
df.num_languages.value_counts().sort_index()
````
