## Załadowanie danych

````python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')
````

Średnia cena

````python
#dodajemy nową kolumnę, która obliczy nam średnią cenę z ceny open i ceny close

df['Average'] = (df.Open + df.Close)/2.0
df.head()
````

Procentowa dzienna zmiana

````python
#sortowanie danych po dacie (czyli indeksie)
df = df.sort_index(ascending = True)
df.head()
````

````python
#metoda shift pozwoli nam przesunąć nasze dane o 1 wiersz
df[['Close']].shift(1)
````

````python
#dodawanie nowej kolumny close_shift
df['Close_shift'] = df[['Close']].shift(1)
````

````python
df.head()
````

````python
#dodawanie nowej kolumny z % zmiany 
df['Daily change %'] = (df.Close / df.Close_shift) - 1
df['Daily change %'] = df['Daily change %'].round(2)
df.head(10)
````

````python
df['Daily_Change %'] = (df.Close / df.Close_shift) - 1
df
````

````python
#usuwanie zbędnych kolumn
del df['Daily change %']
````

````python
df
````

````python
#minimalna i maksymalna wartość z określonej kolumny

df['Daily_Change %'].min()
````

````python
df['Daily_Change %'].max()
````

````python
#wyświetlamy wykres jak kształtowała się zmiana dzienna dla wyceny

df['Daily_Change %'].hist(bins = 100)
````
