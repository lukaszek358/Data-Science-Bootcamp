#Import bibliotek

````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg #interfejs/moduł matplotlib do zdjęć
````

````python
img = mpimg.imread('office.jpg')
img
````

````python
img.shape
````

````python
plt.imshow(img)
plt.show()
````

````python
#wyłączanie naszych osi
plt.imshow(img)
plt.axis('Off')
plt.show()
````

Wyświetlanie siatki obrazów

````python
plt.figure(figsize = (12,10))

plt.subplot(221) # 2 - liczba wierszy w naszej siatce, 2 - liczba kolumn w naszej siatce, 1 - odniesienie do konkretnego wykresu
plt.imshow(mpimg.imread('office.jpg'))
# plt.axis('Off')
# plt.imshow(img)

plt.subplot(222) 
plt.imshow(mpimg.imread('business.jpg'))
# plt.axis('Off')

plt.subplot(223) 
plt.imshow(mpimg.imread('children.jpg'))
# plt.axis('Off')

plt.subplot(224) 
plt.imshow(mpimg.imread('workplace.jpg'))
# plt.axis('Off')

plt.show()
````

Zautomatyzujmy delikatnie nasz kod:

````python
plt.figure(figsize = (12,10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(220 + idx + 1)
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
````

````python
# 1 wiersz i 4 kolumny
plt.figure(figsize = (20,12))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140 + idx + 1)
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
````
