```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set()
```

# Załadowanie danych

## Dataset 'Tips' - wbudowany dataset w bibliotekę Seaborn

```python
df = sns.load_dataset('tips')
df.head(15)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include = 'category').T
```

### Wykres relacji

```python
#wykres rozrzutu
# height - wysokość, aspect - mnożnik do szerokości, width (szerokość) = height * aspect
# hue - podział
sns.relplot(data = df, x = df.total_bill, y = df.tip, height = 5, aspect = 2, hue = df.day)
```

```python
#zmiana wielkości etykiet
sns.set(font_scale = 1.3)
sns.relplot(data = df, x = df.total_bill, y = df.tip, height = 5, aspect = 2, hue = df.day)
```

```python
#dodanie do wykresu podziału w zależności od wielkości grupy (size)

sns.set(font_scale = 1.3)
sns.relplot(data = df, x = 'total_bill', y ='tip', size = 'size', height = 5, aspect = 2, hue = 'size')
```

```python
# podział na zmienne kategoryczne, np. płeć, palących lub nie
# w układzie 1 elementu na wiersz - col_wrap= 1

sns.relplot(data = df, x = df.total_bill, y = df.tip, height = 5, aspect = 2, hue = df.day, col = df.time, col_wrap=1)
```

```python
#row -> ten sam efekt co col_wrap tylko w odniesieniu do jednej ze zmiennych !
sns.relplot(data = df, x = df.total_bill, y = df.tip, height = 5, aspect = 2, hue = df.day, row = df.time)
```

```python
sns.relplot(data = df, x = df.total_bill, y = df.tip, height = 5, aspect = 2, hue = df.day, row = df.time, col = df.smoker)
```

### Wykresy dla danych kategorycznych

```python
sns.catplot(data = df, x = 'day', y = 'total_bill', hue = 'day')
```

```python
sns.catplot(data = df, x = 'day', y = 'total_bill', hue = 'day', kind = 'swarm')
```

```python
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'box')
```

```python
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'violin', height = 5, aspect = 1.5)
```

```python
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'bar', height = 4, aspect = 1.5)
```

```python
#częstość występowania każdego dnia
sns.catplot(data = df, x = 'day', kind = 'count', height = 4, aspect = 1.5)
```

##Dataset 'Titanic'

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['object', 'category', 'bool']).T
```

```python
#częstotliwość rozkładu dla zmiennej 'deck'0
sns.catplot(data = df, x = 'deck', kind = 'count', palette = 'Blues')
```
